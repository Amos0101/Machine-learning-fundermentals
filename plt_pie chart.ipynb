```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exe_vals = [1210,470,500,350,200]
exe_labels = ['Home rent','Food','Phone/internet bill','Car','Other Utilities']
```

```python
plt.axis('equal')
plt.pie(exe_vals, labels = exe_labels,radius = 1, autopct = '%0.1f%%',shadow = False, explode =(0,0,0.1,0,0), startangle = 45)
plt.savefig('piechart.pdf', bbox_inches = 'tight',pad_inches = 2,transparent = True)
plt.show()
```
