```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men = [123,87,89,80,145,134,76,89,141,156,90,98,132]
blood_sugar_women = [133,77,49,70,142,134,71,87,143,136,98,99,137]

plt.xlabel('Sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins= 4, rwidth = 0.95, color = ['green','purple'], label = ['men','women'])
plt.savefig('histogram.pdf', bbox_inches = 'tight',pad_inches = 2,transparent = True)

plt.legend()
```
