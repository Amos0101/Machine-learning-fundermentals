```python
import pandas as pd
df = pd.read_csv("dummy_variables.csv")
df
```

```python
#create dummies(one hot encoding)
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis= 'columns')
merged
```

```python
final = merged.drop(['town','kitui'],axis = "columns")
final
```
