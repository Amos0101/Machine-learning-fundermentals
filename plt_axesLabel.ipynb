```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day = [1,2,3,4,5,6,7]
max_t = [56,58,49,52,47,54,57]
min_t = [36,41,42,34,35,37,32]
avg = [45,43,44,46,42,40,39]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(day,max_t, label='Max')
plt.plot(day,min_t,label='Min')
plt.plot(day,avg,label='avg')

plt.legend(loc = 'lower left', shadow = True, fontsize ='small')
plt.grid()
```
